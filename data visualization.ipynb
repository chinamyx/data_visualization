```python
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

import pandas as pd
from matplotlib import pyplot as plt

x_r = list(range(25,1000))
x=[]
x_rand = []
i = 0

#median  salary for  python developer
yp_r = list(range(10, 100))
yp=[]
yp_rand = []
j = 0

#median salary for all developer
y_r= list(range(100,10000))
y=[]
y_rand  = []
k = 0


for a in x_r:
    if (i <= 10 and a%5 == 0):
        x.append(a)
        i += 1
    
for b in y_r:
    if (j <= 10 and b%5 == 0):
        y.append(b)
        j += 1
    
for c in yp_r:
    if (k <= 10 and c%5 == 0):
        yp.append(c)
        k += 1
    
```

```python
#plot  all developer salary

plt.plot(x, y, color = 'g', linestyle = '--' , label = ' median salary of all developer ')
plt.legend()
plt.xlabel("age")
plt.ylabel('SGD / hr')
plt.title('salary of  all developer')
plt.show()

```

```python
plt.plot(x, yp, color = 'r', linestyle = '-' , label = ' median salary of all python developer ')
plt.legend()
plt.xlabel('age')
plt.ylabel('SGD/ hr')
plt.show()
```

```python
#plot  all developer salary and python developer salary into one single chart

plt.plot(x, y, color = 'g', linestyle = '--' , label = ' median salary of all developer ')
plt.legend()
plt.xlabel("age")
plt.ylabel('SGD / hr')
plt.title('salary of  all developer')


plt.plot(x, yp, color = 'r', linestyle = '-' , label = ' median salary of all python developer ')
plt.legend()
plt.xlabel('age')
plt.ylabel('SGD/ hr')
plt.show()
```
